# 15-6. Two Dice: 
Create a simulation showing what happens when you roll two eight-sided dice 1,000 times. Try to picture what you think the visualization will look like before you run the simulation, then see if your intuition was correct. Gradually increase the number of rolls until you start to see the limits of your systemâ€™s capabilities.

```python
import plotly.express as px
from random import randint
class Die:
    """A class representing a single die."""
    def __init__(self, num_sides=6): 
         
      """Assume a six-sided die."""

      self.num_sides = num_sides
    def roll(self):
          """"Return a random value between 1 and number of sides."""
          return randint(1, self.num_sides)
```

```python
# Create two D8 dice.
die_1 = Die(num_sides=8)
die_2 = Die(num_sides=8)

# Roll the dice 1,000 times and store the sums.
results = []
num_rolls = 1000  # Adjust this value to test system capabilities.
for _ in range(num_rolls):
    result = die_1.roll() + die_2.roll()
    results.append(result)

# Analyze the results.
frequencies = []
poss_results = range(2, die_1.num_sides + die_2.num_sides + 1)  # Possible sums range from 2 to 16.
for value in poss_results:
    frequency = results.count(value)
    frequencies.append(frequency)

# Visualize the results.
fig = px.bar(
    x=list(poss_results),
    y=frequencies,
    labels={'x': 'Sum of Two D8 Rolls', 'y': 'Frequency'},
    
    title=f"Results of Rolling Two D8 Dice {num_rolls} Times"
)
fig.update_layout(xaxis_dtick=1)
fig.show()
```

## Die-Rolling Visualization with Matplotlib

```python
import random
import matplotlib.pyplot as plt

# Define the Die class
class Die:
    """A class to represent a single die."""
    
    def __init__(self, num_sides=6):
        """Initialize the die with a number of sides."""
        self.num_sides = num_sides

    def roll(self):
        """Return a random value between 1 and the number of sides."""
        return random.randint(1, self.num_sides)

# Create two D6 dice.
die_1 = Die()
die_2 = Die()

# Roll the dice and store the sums.
results = []
num_rolls = 1000  # Number of rolls
for roll_num in range(num_rolls):
    result = die_1.roll() + die_2.roll()  # Sum of both dice
    results.append(result)

# Analyze the results.
max_result = die_1.num_sides + die_2.num_sides  # Maximum result is 12 (6+6)
frequencies = [results.count(value) for value in range(2, max_result + 1)]  # Possible sums range from 2 to 12

# Visualize the results using Matplotlib.
plt.figure(figsize=(10, 6))
plt.bar(range(2, max_result + 1), frequencies)
plt.title(f"Results of Rolling Two D6 Dice {num_rolls} Times")
plt.xlabel("Sum of Two D6 Rolls")
plt.ylabel("Frequency")
plt.xticks(range(2, max_result + 1))  # Set x-ticks for better readability
plt.show()
```

## Random Walk Visualization with Plotly

```python
import random
import plotly.graph_objects as go

class RandomWalk:

    """A class to generate random walks."""
    
    def __init__(self, num_points=5000):
        """Initialize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0, 0).
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the walk."""
        while len(self.x_values) < self.num_points:
            x_step = self.get_step()
            y_step = self.get_step()
            if x_step == 0 and y_step == 0:
                continue
            
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
            self.x_values.append(x)
            self.y_values.append(y)

    def get_step(self):
        """Determine the direction and distance for each step."""
        direction = random.choice([1, -1])
        distance = random.choice([0, 1, 2, 3, 4])
        return direction * distance
```

```python

# Create a random walk.
rw = RandomWalk()
rw.fill_walk()

# Visualize the random walk using Plotly.
fig = go.Figure()

# Add the walk to the figure.
fig.add_trace(go.Scatter(x=rw.x_values, y=rw.y_values, mode='lines', name='Random Walk'))

# Emphasize the start and end points.
fig.add_trace(go.Scatter(x=[0], y=[0], mode='markers', marker=dict(size=10, color='green'), name='Start',))
fig.add_trace(go.Scatter(x=[rw.x_values[-1]], y=[rw.y_values[-1]], mode='markers', marker=dict(size=10, color='red'), name='End'))

# Update layout.
fig.update_layout(title='Random Walk Visualization',
                  xaxis_title='X Values',
                  yaxis_title='Y Values',
                  showlegend=True)

fig.show()
```
